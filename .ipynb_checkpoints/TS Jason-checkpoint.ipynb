```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
from pandas import Series
from datetime import datetime
```

```python
data=pd.read_csv("dataNew.csv")
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.set_index('Date',inplace=True)
data.head()
```

```python
ts=data['Open']
ts.head()
```

```python
plt.figure(figsize=(15,10))
ts.plot()
```

```python
# AR example
from statsmodels.tsa.ar_model import AR
from random import random
# fit model
model = AR(ts)
model_fit = model.fit()
# make prediction
yhat = model_fit.predict(len(ts), len(ts))
print(yhat)
```

```python
# MA example
from statsmodels.tsa.arima_model import ARMA
from random import random
# fit model
model = ARMA(ts, order=(0, 1))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(ts), len(ts))
print(yhat)
```

```python
# ARMA example
from statsmodels.tsa.arima_model import ARMA
from random import random
# fit model
model = ARMA(ts, order=(2, 1))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(ts), len(ts))
print(yhat)
```

```python
# ARIMA example
from statsmodels.tsa.arima_model import ARIMA
from random import random
# fit model
model = ARIMA(ts, order=(1, 2, 1))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(ts), len(ts), typ='levels')
print(yhat)
```
