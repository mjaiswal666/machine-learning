```python
#Now we use adfuller method
from statsmodels.tsa.stattools import adfuller
```

```python
def test_stationarity(timeseries):

    #As it is difficult to find the mean or the standard deviation
    #rolmean = pd.rolling_window(timeseries, window=12)  # As we have data of 12 months
    rolmean = timeseries.rolling(window=7).mean()  # As we have data of 12 months
    #rolstd = pd.rolling_std(timeseries, window=12) #As we have data of 12 months
    rolstd = timeseries.rolling( window=7).std() #As we have data of 12 months
    #Now we plot the rolling statistics

    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    # Perform Dickry Fuller test
    print ('Result of Dickey-Fuller test')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical value(%s)' % key] = value
    print (dfoutput)
```

```python
test_stationarity(cl)
```

```python
cl_log=np.log(cl)
plt.plot(cl_log)
```

```python
cl_log
cl_log_diff =cl_log.diff(1)
cl_log_diff.dropna(inplace=True)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition=seasonal_decompose(cl_log)
```

```python
trend=decomposition.trend
trend
plt.plot(trend)
```

```python
seasonality=decomposition.seasonal
plt.plot(seasonality)
```

```python
residual=decomposition.resid
plt.plot(residual)

cl_log_decompose = residual

cl_log_decompose.dropna(inplace=True)
```

```python
plt.subplot(411)

plt.plot(cl_log,label='Original')

plt.legend(loc='best')

plt.subplot(412)

plt.plot(trend,label='Trend')

plt.legend(loc='best')

plt.subplot(413)

plt.plot(seasonality,label='Seasonarity')

plt.legend(loc='best')

plt.subplot(414)

plt.plot(residual,label='Residual')

plt.legend(loc='best')

plt.tight_layout()

plt.show()
```

```python
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(cl_log_diff, nlags=20)

lag_pacf = pacf(cl_log_diff, nlags=20, method='ols')
```

```python
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96 / np.sqrt(len(cl_log_diff)), linestyle='--', color='gray')  # lowwer
plt.axhline(y=1.96 / np.sqrt(len(cl_log_diff)), linestyle='--', color='gray')  # upper
plt.title('Autocorrelation Function')
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96 / np.sqrt(len(cl_log_diff)), linestyle='--', color='gray')
plt.axhline(y=1.96 / np.sqrt(len(cl_log_diff)), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
plt.show()
```

```python
import warnings
from pandas import read_csv
from pandas import datetime
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python

model = ARIMA(cl_log, order=(5, 1, 5))

result_ARIMA = model.fit(disp=-1)

plt.plot(cl_log_diff)

plt.plot(result_ARIMA.fittedvalues, color='red')

plt.title('ARIMA RSS:%.4f' % sum(result_ARIMA.fittedvalues - cl_log_diff) ** 2)

plt.show()
```

```python
result_ARIMA.summary()
```

```python
model_arima = ARIMA(cl,order=(0, 2 , 0))
model_arima_fit = model_arima.fit()
print(model_arima_fit.aic)
predictions= model_arima_fit.forecast(steps=10)[0]
predictions
```

```python
predictions_ARIMA_diff = pd.Series(result_ARIMA.fittedvalues, copy=True)



predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()



predictions_ARIMA_log = pd.Series(cl_log.ix[0], index=cl_log.index)

predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value=0)

# print predictions_ARIMA_log.head()



predictions_ARIMA = np.exp(predictions_ARIMA_log)

plt.plot(cl)

plt.plot(predictions_ARIMA)

plt.title('predictions_ARIMA RMSE: %.4f' % np.sqrt(sum((predictions_ARIMA - cl) ** 2) / len(cl)))

plt.show()
```
